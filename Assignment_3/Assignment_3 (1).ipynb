```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\HP\Downloads\advertising.csv"
df = pd.read_csv(file_path)

# Display basic info
print(df.info())
print(df.head())  # Show the first few rows
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["TV"], y=df["Sales"])
plt.xlabel("TV Advertisement Spend")
plt.ylabel("Sales")
plt.title("TV Ad Spend vs Sales")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Radio", y="Sales", data=df)
plt.xlabel("Radio Advertisement Spend")
plt.ylabel("Sales")
plt.title("Radio Ad Spend vs Sales")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df["Newspaper"])
plt.xlabel("Newspaper Advertisement Spend")
plt.title("Box Plot of Newspaper Ad Spend")
plt.show()
```

```python
ad_spend = [df["TV"].sum(), df["Radio"].sum(), df["Newspaper"].sum()]
labels = ["TV", "Radio", "Newspaper"]
colors = ["blue", "green", "red"]

plt.figure(figsize=(7,7))
plt.pie(ad_spend, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Advertisement Spend Distribution")
plt.show()
```

```python
if "Day" in df.columns:
    plt.figure(figsize=(8,6))
    plt.plot(df["Day"], df["Sales"], marker='o', linestyle='-', color='purple')
    plt.xlabel("Day")
    plt.ylabel("Sales")
    plt.title("Sales Over Time")
    plt.grid()
    plt.show()
```
