```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"F:\VIIT\admission.csv")

# Display first few rows to understand the data
print(df.head())

# ðŸ”¹ Box Plot: GRE & TOEFL Score Distribution
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[["GRE Score", "TOEFL Score"]])
plt.title("Box Plot of GRE & TOEFL Scores")
plt.show()


```

```python
print(df.columns)
```

```python
# ðŸ”¹ Bar Plot: Average GRE Score by University Rating
plt.figure(figsize=(8, 4))
df.groupby("University Rating")["GRE Score"].mean().plot(kind="bar", color="green")
plt.xlabel("University Rating")
plt.ylabel("Average GRE Score")
plt.title("Bar Plot: Average GRE Score by University Rating")
plt.xticks(rotation=0)
plt.show()
```

```python
# ðŸ”¹ Scatter Plot
plt.figure(figsize=(6, 4))
plt.scatter(df["GRE Score"], df["TOEFL Score"], color='blue', alpha=0.5)
plt.xlabel("GRE Score")
plt.ylabel("TOEFL Score")
plt.title("Scatter Plot: GRE vs TOEFL Score")
plt.grid(True)
plt.show()
```

```python
# ðŸ”¹ Line Chart: GRE Score vs TOEFL Score
plt.figure(figsize=(8, 5))
plt.plot(df["GRE Score"], df["TOEFL Score"], marker='o', linestyle='-', color='b', alpha=0.6)
plt.xlabel("GRE Score")
plt.ylabel("TOEFL Score")
plt.title("Line Chart: GRE Score vs TOEFL Score")
plt.grid(True)
plt.show()
```

```python
# ðŸ”¹ Pie Chart: Distribution of University Ratings
plt.figure(figsize=(6, 6))
df["University Rating"].value_counts().plot(kind="pie", autopct='%1.1f%%', colors=["blue", "green", "red", "purple", "orange"])
plt.title("Pie Chart: University Rating Distribution")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ”¹ Set figure size
plt.figure(figsize=(12, 6))

# ðŸ”¹ Box Plot for different columns
sns.boxplot(data=df[["GRE Score", "TOEFL Score", "SOP", "LOR ", "CGPA", "Research"]])

# ðŸ”¹ Labels & Title
plt.xticks(rotation=15)  # Rotate labels for readability
plt.title("Box Plot for GRE, TOEFL, SOP, LOR, CGPA, and Research")
plt.grid(True)

# ðŸ”¹ Show the plot
plt.show()
```
